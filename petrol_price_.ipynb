```python
# A machine learning library used for linear regression
from sklearn.linear_model import LinearRegression
# numpy and pandas will be used for data manipulation
import numpy as np
import pandas as pd
# matplotlib will be used for visually representing our data
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
from google.colab import files 
   
uploaded = files.upload()
import io 
```

```python
data = pd.read_csv(io.BytesIO(uploaded['petrol.csv'])) 
```

```python
data.Chennai.plot(figsize=(10,5))
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['Date']=label_encoder.fit_transform(data['Date'])
data['Date'].unique()
```

```python
# Dropping the NaN values
data = data.dropna()

# Initialising X and assigning the two feature variables
X = data[['Date','Chennai']]

# Getting the head of the data
X.head()
```

```python
# Setting-up the dependent variable
y = data['Mumbai']

# Getting the head of the data
y.head()
```

```python
# Setting the training set to 80% of the data
training = 0.8
t = int(training*len(data))

# Training dataset
X_train = X[:t]
y_train = y[:t]

# Testing dataset
X_test = X[t:]
y_test = y[t:]
```

```python
model = LinearRegression().fit(X_train,y_train)
```

```python
predicted_price = model.predict(X_test)
predicted_price = pd.DataFrame(predicted_price,index=y_test.index,columns = ['price'])
predicted_price.plot(figsize=(10,5))
y_test.plot()
plt.legend(['Predicted Price','Actual Price'])
plt.ylabel("petro Prices: chennai-mumbai")
plt.show()
```

```python
from pandas import Series;  from numpy.random import randn
```

```python
from sklearn.metrics import mean_squared_error
MSE=mean_squared_error(y_test,predicted_price)
print('Mean Squared Error is:', MSE)
```

```python
import math
RMSE = math.sqrt(MSE)
print('Root Mean Squared Error is:', RMSE)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predicted_price)
print('R-Squared is:', r2)
```
